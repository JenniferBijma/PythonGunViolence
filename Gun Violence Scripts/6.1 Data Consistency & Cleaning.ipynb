```python
# Dropping unnecessary Columns
df_gun.drop(columns = {'incident_url', 'source_url', 'participant_name', 'sources', 'notes', 'incident_url_fields_missing', 'location_description'}, inplace = True)
```

```python
# Finding Missing values
df_gun.isnull().sum()
```

```python
# Dropping columns with too many missing values
df_gun.drop(columns = {'gun_stolen', 'gun_type', 'n_guns_involved', 'participant_age', 'participant_relationship'}, inplace = True)
```

```python
#Check for Mixed Types
for col in df_gun.columns.tolist():
  weird = (df_gun[[col]].applymap(type) != df_gun[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_gun[weird]) > 0:
    print (col)
```

```python
# Changing Data Type of mixed columns
df_gun[['address', 'incident_characteristics', 'participant_age_group', 'participant_gender', 'participant_status', 'participant_type']] = df_gun[['address', 'incident_characteristics', 'participant_age_group', 'participant_gender', 'participant_status', 'participant_type']].astype('str')
```
